```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('namesbystate_ranks_counts.csv')
```

```python
latest_data = data[data['year'] == data['year'].max()]
```

```python
latest_data = latest_data.sort_values(['state', 'sex', 'rank'])
```

```python
def top_N_ex_state_names(sex, state, N):
    top_ten_ex = latest_data[
        (latest_data['rank'] < N) &
        (latest_data['sex'] == sex) &
        (latest_data['state'] != state)
    ]
    return set(top_ten_ex['name'])
```

```python
top_ten_ex_ohio = top_N_ex_state_names('M', 'OH', 10)
top_ten_ex_ohio
```

```python
latest_data[
    (latest_data['state'] == 'OH') &
    (latest_data['sex'] == 'M') &
    ~latest_data['name'].str.fullmatch('|'.join(top_ten_ex_ohio))]
```

```python
def top_N_distinctive(sex, state, N):
    return latest_data[
        (latest_data['state'] == state) &
        (latest_data['sex'] == sex) &
        ~latest_data['name'].str.fullmatch('|'.join(top_N_ex_state_names(sex, state, N)))].iloc[:N]
```

```python
top_N_distinctive('M', 'OH', 10)
```
